```python
import numpy as np
import matplotlib.pyplot as plt

iris = np.genfromtxt('iris_2D_demo.csv',
                     dtype=None,
                     delimiter=',',
                     skip_header=1)

x_data = iris[:, :2]
y_data = iris[:, 2]

x_data_0 = x_data[y_data == 0]
x_data_1 = x_data[y_data == 1]

plt.figure(figsize=(6, 4))
plt.scatter(x_data_0[:, 0], x_data_0[:, 1], color='b', label='0')
plt.scatter(x_data_1[:, 0], x_data_1[:, 1], color='r', label='1')
plt.legend()
```

```python
N = x_data.shape[0]
intercept = np.ones((N, 1))
x_data = np.c_[intercept, x_data]

print('x_data \n', x_data)
print('y_data \n', y_data)
```

image.png

```python
def sigmoid_function(z):
    return 1 / (1 + np.exp(-z))

def predict(X, theta):
    return sigmoid_function(np.dot(X.T, theta))

def loss_function(y_hat, y):
    return (-y*np.log(y_hat) - (1 - y)*np.log(1 - y_hat))

def conpute_gradient(X, y_hat, y):
    return X*(y_hat - y)

def update(theta, lr, gradient):
    return theta - lr*gradient
```

image.png

```python
# initialization
lr = 0.01
max_epoch = 1

theta = np.array([0.1, 0.5, -0.1])
losses = []

for epoch in range(max_epoch):
    for i in range(N):    # N = x_data.shape[0]
        X = x_data[i]
        y = y_data[i]
        print(f'X: {X}')
        print(f'y: {y}')

        # compute output (y_hat)
        y_hat = predict(X, theta)
        print(f'y_hat: {y_hat}')

        # compute loss
        loss = loss_function(y_hat, y)
        losses.append(loss)
        print(f'loss: {loss}')

        # compute mean of gradient
        gradient = conpute_gradient(X, y_hat, y)
        print(f'gradient: {gradient}')

        # update
        theta = update(theta, lr, gradient)
        print(f'theta: {theta}\n\n')
```
