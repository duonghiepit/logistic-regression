```python
iris = np.genfromtxt('iris_1D.csv',
                     dtype=None,
                     delimiter=',',
                     skip_header=1)

x_data = iris[:, 0]
y_data = iris[:, 1]

print(x_data)
print(y_data)
```

```python
def sigmoid_function(z):
    return 1 / (1 + np.exp(-z))

def predict(x, w, b):
    z = x*w + b
    print(f'z: {z}')

    return sigmoid_function(z)

def compute_loss(y_hat, y):
    return -y*np.log(y_hat) - (1-y)*np.log(1 - y_hat)

def compute_gradient(x, y_hat, y):
    db = y_hat - y
    dw = x*(y_hat -y)

    return db, dw

def update(b, w, lr, db, dw):
    w = w - lr*dw
    b = b - lr*db

    return b, w
```

```python
lr = 0.01
num_epoch = 1

w = -0.1
b = 0.1

# for monitoring
losses = []

for _ in range(num_epoch):
    for i in range(len(x_data)):
        # get data
        x = x_data[i]
        y = y_data[i]
        print('x, y:', x, y)

        # predict
        y_hat = predict(x, w, b)
        print(f'y_hat: {y_hat}')

        # compute loss
        loss = compute_loss(y_hat, y)
        losses.append(loss)
        print(f'loss: {loss}')

        # compute gradient
        db, dw = compute_gradient(x, y_hat, y)
        print(f'db: {db}, dw: {dw}')

        # update
        b, w = update(b, w, lr, db, dw)
        print(f'b_new: {b}, w_new: {w}')
        print('-------')
```
